```python
# Import necessary libraries
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler

# Load and normalize the Iris dataset
iris = datasets.load_iris()
data = iris.data
target = iris.target

# Normalize data
scaler = MinMaxScaler()
data = scaler.fit_transform(data)

# Define SOM dimensions and parameters
som_size = (20, 20)
som = MiniSom(som_size[0], som_size[1], data.shape[1], sigma=1.0, learning_rate=0.5)

# Initialize SOM weights
som.random_weights_init(data)

# Train the SOM with 100 iterations
som.train_random(data, 100)

# Plot the SOM with the weight of the winner neurons (initial training with 100 iterations)
plt.figure(figsize=(10, 10))
plt.pcolor(som.distance_map().T, cmap='coolwarm')
plt.colorbar(label='Distance')
markers = ['o', 's', 'D']  # Circle, square, and diamond markers
colors = ['C0', 'C1', 'C2']

for i, x in enumerate(data):
    w = som.winner(x)
    plt.plot(w[0] + 0.5, w[1] + 0.5, markers[target[i]], markerfacecolor='None',
             markeredgecolor=colors[target[i]], markersize=12, markeredgewidth=2)

plt.title('SOM - 100 Iterations')
plt.show()

# Retrain the SOM with 50,000 iterations
som.train_random(data, 50000)

# Plot the SOM with the weight of the winner neurons (after retraining with 50,000 iterations)
plt.figure(figsize=(10, 10))
plt.pcolor(som.distance_map().T, cmap='coolwarm')
plt.colorbar(label='Distance')

for i, x in enumerate(data):
    w = som.winner(x)
    plt.plot(w[0] + 0.5, w[1] + 0.5, markers[target[i]], markerfacecolor='None',
             markeredgecolor=colors[target[i]], markersize=12, markeredgewidth=2)

plt.title('SOM - 50,000 Iterations')
plt.show()
```
